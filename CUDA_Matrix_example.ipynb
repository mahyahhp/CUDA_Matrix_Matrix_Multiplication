```python
# d_M=|d_M0      d_M1      d_M2      d_M3 |     d_N=|d_N0      d_N1      d_N2      d_N3 |
#     |d_M4      d_M5      d_M6      d_M7 |         |d_N4      d_N5      d_N6      d_N7 |
#     |d_M8      d_M9      d_M10     d_M11|         |d_N8      d_N9      d_N10     d_N11|
#     |d_M12     d_M13     d_M14     d_M15|         |d_N12     d_N13     d_N14     d_N15|

# d_p=d_M*d_N
dic={}
by=[[-1]*2 for i in range(2)]
bx=[[-1]*2 for i in range(2)]
ty=[[-1]*2 for i in range(2)]
tx=[[-1]*2 for i in range(2)]
Row=[[-1]*2 for i in range(2)]
Col=[[-1]*2 for i in range(2)]
pvalue=[["0"]*2 for i in range(2)]

width=4
TILE_WIDTH=2 #this is equals to blockDim.x and blockDim.y
for blockIdx_y in range(2):
    for blockIdx_x in range(2):
        print(f"")
        print(f"")
        print(f"***********************************************")
        print(f"***********************************************")
        print(f"Block({blockIdx_y},{blockIdx_x})")
        for ph in range(round(width/TILE_WIDTH)):
            for threadIdx_y in range(2):
                for threadIdx_x in range(2):
                    print(f"---------------------------------------------")
                    print(f"thread({threadIdx_y},{threadIdx_x})")
                    by[threadIdx_y][threadIdx_x]=blockIdx_y
                    bx[threadIdx_y][threadIdx_x]=blockIdx_x
                    ty[threadIdx_y][threadIdx_x]=threadIdx_y
                    tx[threadIdx_y][threadIdx_x]=threadIdx_x
                    Row[threadIdx_y][threadIdx_x]=by[threadIdx_y][threadIdx_x]*TILE_WIDTH+ty[threadIdx_y][threadIdx_x]
                    Col[threadIdx_y][threadIdx_x]=bx[threadIdx_y][threadIdx_x]*TILE_WIDTH+tx[threadIdx_y][threadIdx_x]
                    print(f"row={Row[threadIdx_y][threadIdx_x]}")
                    print(f"col={Col[threadIdx_y][threadIdx_x]}")
                    pvalue[threadIdx_y][threadIdx_x]="0"
                    print(f"ph={ph}")

                    key="Mds["+str(ty[threadIdx_y][threadIdx_x])+"]["+str(tx[threadIdx_y][threadIdx_x])+"]"
                    item="d_M["+str(Row[threadIdx_y][threadIdx_x]*width+ph*TILE_WIDTH+tx[threadIdx_y][threadIdx_x])+"]"
                    dic.update({key:item})
                    key="Nds["+str(ty[threadIdx_y][threadIdx_x])+"]["+str(tx[threadIdx_y][threadIdx_x])+"]"
                    item="d_N["+str((ph*TILE_WIDTH+ty[threadIdx_y][threadIdx_x])*width+Col[threadIdx_y][threadIdx_x])+"]"
                    dic.update({key:item})

                    d_M_index=Row[threadIdx_y][threadIdx_x]*width+ph*TILE_WIDTH+tx[threadIdx_y][threadIdx_x]
                    d_N_index=(ph*TILE_WIDTH+ty[threadIdx_y][threadIdx_x])*width+Col[threadIdx_y][threadIdx_x]
                    print(f"Mds[{ty[threadIdx_y][threadIdx_x]}][{tx[threadIdx_y][threadIdx_x]}]=d_M[{d_M_index}]")
                    print(f"Nds[{ty[threadIdx_y][threadIdx_x]}][{tx[threadIdx_y][threadIdx_x]}]=d_N[{d_N_index}]")
                    # print(f"__synchthreads()")
            for threadIdx_y in range(2):
                for threadIdx_x in range(2):     
                    print(f"---------------------------------------------")
                    print(f"thread({threadIdx_y},{threadIdx_x})")
                    for k in range(TILE_WIDTH):
                        Mds_tex="Mds["+str(ty[threadIdx_y][threadIdx_x])+"]["+str(k)+"]"
                        Nds_tex="Nds["+str(k)+"]["+str(tx[threadIdx_y][threadIdx_x])+"]"
                        print(f"k={k}")
                        # print(f"pvalue=pvalue+Mds[{ty[threadIdx_y][threadIdx_x]}][{k}]*Nds[{k}][{tx[threadIdx_y][threadIdx_x]}]      or     pvalue=pvalue+{dic[Mds_tex]}*{dic[Nds_tex]}")
                        pvalue[threadIdx_y][threadIdx_x]=str(pvalue[threadIdx_y][threadIdx_x])+"+"+dic[Mds_tex]+"*"+dic[Nds_tex]
                        print(f"pvalue[{threadIdx_y}][{threadIdx_x}]={pvalue[threadIdx_y][threadIdx_x]}")
        for threadIdx_y in range(2):
            for threadIdx_x in range(2):
                print(f"---------------------------------------------")
                print(f"d_p[{Row[threadIdx_y][threadIdx_x]*width+Col[threadIdx_y][threadIdx_x]}]={pvalue[threadIdx_y][threadIdx_x]}")




```
